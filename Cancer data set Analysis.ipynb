The data type in the columns is obejct thus the selected columns may contain both textual and numeric data. There is need to convert the data into numeric inorder to make a numeric analysis

```python
# Convert selected numeric columns to numeric data type
numeric_columns = [
     'Cancers attributable to infections\nProportion of cancers attributable to infections (%), by country',
     'Obesity prevalence male\nInternational variation in the prevalence of obesity, 2016',
     'Obesity prevalence female\nInternational variation in the prevalence of obesity, 2016',
     'Most common cancer cases worldwide, females\n2018',
     'Average births per woman\n2010-2015',
     'Breastfeeding at 12 months\nPercent (%) of children who receive any breast milk at 12 months of age',
     'Most common cancer cases worldwide, males\n2018',
     'Most common cancer deaths worldwide, females\n2018',
     'Most common cancer deaths worldwide, males\n2018',
     'Lung cancer incidence rates, male\nAge-standardized rate (world) per 100,000, all ages, 2018',
     'Lung cancer incidence rates, female\nAge-standardized rate (world) per 100,000, all ages, 2018'
]
selected_fp = selected_fp.copy()
selected_fp[numeric_columns] = selected_fp[numeric_columns].apply(pd.to_numeric, errors='coerce')
selected_fp[numeric_columns]
```

It represents the percentage represented by the column in each country. For example for column 2 row 0, it represents that 16% of cancer in that country is caused by infections.

```python
# 1.Check for any missing values
missing_values = selected_fp.isnull().sum()
missing_values
```

```python
# 2.Confirm for any duplicates
duplicates = selected_fp.duplicated().sum()
duplicates
```

```python
# 3.Remove missing values
cleaned_fp = selected_fp.dropna()
cleaned_fp

```

```python
# Check for any remaining missing values
missing_values = cleaned_fp.isnull().sum()
missing_values
```
